- 스웨덴과 영국 모두 공유 자전거를 출퇴근 시간대에 많이 사용

### 12주차

process 1: 자전거 수요 예측에 대한 사전 methodology 조사

자전거 수요 예측 변수
날씨, 대기오염수준, 자전거 도로 인프라, 코로나19 발생

자전거 수요 예측 알고리즘
- SVR(Support Vector Regression), LR(Linear Regression), LSTM -> RMSE로 비교
- 다른 논문들: KNN, MRGNN(인접 지하철역 정보와 관계를 설명하는 다중관계 그래프 신경망) 
-> 강화 학습을 이용해서 자전거 수요를 예측한 선행 논문이 없다. 
2021년도에 발표된 논문에 자전거가 아닌 자동차 교통량 예측을 강화학습에 적용한 사례가 있었는데, 이 역시 일반적인 모델에 데이터가 부족할 때 강화학습을 추가한 형태였다. 이 논문에 의하면 21년도 7월까지 강화학습을 통해서 교통량을 예측한 사례가 없다고 한다. 그러나 공유 자전거 불균형문제를 재조정하는데 있어서 강화학습을 사용한 사례는 여럿 있다.

강화 학습을 통해서 교통량을 예측하는 사례가 없는 이유 조자를 하였는데, 강화 학습은 기존의 데이터를 가지고 최적의 의사결정을 내리는 모델로 사용이 되고, 우리가 하는 연구처럼 과거의 데이터를 토대로 자전거 수요량을 예측하는 것이면, 다른 LSTM, SVM이 더 적합하다는 것을 찾을 수 있었다.


자전거 수요 예측을 위한 rule-based regression predictive model
  - 방법론: CUBIST, Regularized Random Forest, Classification and Regression Tree, KNN, Conditional Inference Tree
   - __CUBIST__ : 가장 높은 정확도를 나타냄
     - 목표 변수와 관련된 다양한 예측 변수들 간의 비선형 관계를 탐색하기 위해 Decision Tree와 Regression 모델을 결합한 알고리즘으로 구성
     - Decision Tree에서 앙상블 구성 : 입력 변수들을 기반으로 여러 개의 Decision Tree를 독립적으로 학습한다
     - 예측 규칙 결합 : Decision Tree에서 생성된 예측 규칙들을 결합하여 종합적인 예측을 수행 -> 특정 조건에 따라 예측을 수행하는데 사용된다
     - 선형 회귀 모델 학습 : 예측 규칙들을 기반으로 선형 회귀 모델을 학습한다 -> 비선형 관계와 선형 관계를 모두 고려하여 예측 진행
   - Regularized Random Forest
     - Random Forest 모델에 정규화 기법을 적용하여 변형된 버전
     - 정규화 : overfitting 방지 및 일반화 성능 향상
     - 가지치기 : Tree를 잘라내어 불필요한 가지 제거, Tree의 깊이 제한
     - 변수 중요도 기반 선택 : 변수 중요도를 기준으로 일부 변수를 선택하여 다른 변수를 제외
     - 부분 샘플링 : 데이터의 일부만을 사용하여 각 Decision Tree 학습
   - Classification and Regression Tree (CART)
     - 목표 변수 사이의 비선형 관계를 학습하여 예측 수행
     - Classification or Regression 중 하나에 특화되어 진행
     - Single Decision Tree를 사용하여 예측을 수행
   - KNN(K-Nearest Neighbors)
     - 이웃의 개수를 K로 설정하여 가장 가까운 K들을 찾는다
     - 선택한 이웃의 label 혹은 연속적인 값을 기반으로 주어진 데이터의 label을 예측
   - Conditional Inference Tree
     - 분할 규칙을 학습하기 위해 조건부 추론 기법을 사용
     - 데이터의 특징과 목표 변수 사이의 관계를 조건부 추론 테스트를 통해 검증하고, 가장 유의한 특징을 기준으로 데이터를 분할
     - 분할된 Tree를 기반으로 예측을 수행
     - label과 regression 모두 사용

- Process1. GCN + FCL

  - Pipeline1<br>
    GCN 모델을 이용해서 각 거치소의 수요량을 예측할 수 있다.

  - 기대 효과<br>
    거치대별 자전거 수요 예측을 통해 자전거 대여 회사나 도시 관리 기관 등이 자원을 효율적으로 할당할 수 있도록 도와줍니다. 예를 들어, 수요가 많을 것으로 예상되는 지역에 자전거를 미리 공급하여 수급 부족 문제를 방지하거나, 반대로 수요가 낮을 것으로 예상되는 지역의 자전거를 다른 지역으로 이동시켜 자원 이용을 최적화할 수 있습니다. (그런데 이것은 통계학적인 방법으로 접근하는 것에 비해 어떤 장점이 있는지 모르겠다.) (만약 우리가 생각하고 있는 변수들이 변한다고 가정했을 때(지하철역 혹은 버스정류장 혹은 기타상업시설의 입지) 그런 공간적인 데이터를 수정했을 때 동적으로 변하는 상황에 따라 달라지는 자전거 수요량을 예측할 수 있다는 것에 대해서 강점이 있을 것이다.

  - 방법<br>
    GCN에는 두가지 행렬을 인풋으로 받는다. 각 노드들의 특징을 설명하는 행렬, 노드 간의 관계를 설명하는 행렬.

    각 거치소의 특징을 설명하는 행렬과 각 거치소의 관계를 설명하는 행렬을 인풋으로 한다.
    각 거치소의 특징이라면 공간적인 정보이다. 예를 들면, 인근 지하철역 개수, 버스정류장 개수, 학교 수, 상업시설 수 등등이다. 
    그리고 각 거치소의 관계 행렬(인접 행렬)은 거치소 출발장소와 도착장소가 있는 데이터를 사용하여 나타낼 것이다. 예를 들면 1번 거치소에서 3번 거치소로 가는 자전거의 수가 많으면 (1,3)에 큰 수를 기입하고, 3번 거치소에서 2번 거치소로 가는 자전거의 수가 적으면 (3,2)의 수를 적게 한다.

- Process2. DeepAR

  - 기대 효과<br>
    시간에 대한 데이터에 집중적으로 포커스하여 모델링을 진행할 수 있다. 시간에 따라 변할 수 있는 데이터인 유동 인구수, 날씨 및 기상청 데이터, 특징적인 시간(공휴일, 주말 등)을 독립변수로 설정하여 각 시간 혹은 일별로 자전거의 수요를 예측할 수 있다. 만약 서울특별시 내의 행정구, 행정동별로 구분하여 프로젝트를 진행한다면 특정 위치(정거장)의 수요를 예측할 수 있다. 또한 DeepAR의 특징인 확률론적인 분포를 보여준다는 것을 토대로 다양한 불확실성을 고려하여 진행할 수 있다.
  - 한계점<br>
    공간적인 데이터를 다룰 수 없는 모델이므로, 모든 데이터를 행정구, 행정동 별로 분류해야 한다.

  - 방법:RNN(LSTM)<br>
    - 1.	범주형 데이터의 embedding<br>
      수치로 나타나지 않는 범주형 데이터 (요일, 계절, 휴일 등)을 embedding 과정을 거쳐 벡터로 전환한다.
    - 2.	재귀 신경망 (RNN)<br>
     과거의 관측값을 현재의 예측 값을 계산하는데 활용하는 모델이며, vanishing gradient 문제를 해결하기 위해 LSTM 기법을 적용하여 신경망을 구성할 수 있다. LSTM의 RNN cell은 각각의 시점으로 구성되어 있고, 이전 시점의 예측이 다음 시점의 input으로 들어갈 수 있다. LSTM에 들어가는 시점별 변수는 벡터의 형태로 모델에 input되며, LSTM에서는 해당 시간 단계에서 계산된 값을 DeepAR에서는 시계열의 예측 값을 출력한다.
    - 3.	훈련 (Training)<br>
    - 4.	예측 단계<br>
     훈련이 완료된 DeepAR 모델은 새로운 시계열 데이터에 대한 예측을 진행할 수 있다. 이 때, 시계열 데이터의 이전 과거 값을 입력으로 받아서 미래의 시계열 값을 생성한다.

- Process 3. Meanshift clustering + RFC 

  - 기대효과: 클러스터에 영향을 끼치는 중요한 요인 분석.<br>

  - 한계점<br> 

  - 방법: Meanshift clustering + RFC<br> 

    - 1.	Meanshift clustering<br>
      09:00의 자전거의 위치를 기준으로 자전거의 클러스터링 영역을 식별
      Bandwidth를 300m, 500m만 고려하고 이때 식별되는 영역은 그리드에서 각각 500m*500m,1000m*1000m.
    이 중 더 촘촘히 짜여지는 300m의 결과인 492개
    - 2.	Buffer 분석<br>
      자전거 집결지의 범위를 구하기 위해 버퍼 분석을 제안. 버퍼는 공간 대상의 일종의 영향 범위 또는 서비스 범위로 점, 선, 면 개체를 중심으로 자동적으로 설정되는 일정 폭의 다각형을 의미. 모든 클러스터 중심점을 기준으로 300미터의 버퍼를 ArcGIS로 설정하여 자전거 집결지의 도킹리스 PBS 지표와 영향력 지수를 산출.
    - 3.	K-means<br>
      자전거 클러스터 영역의 가용 자전거 수와 가용 자전거 수 변동 계수를 계산한 후, K-평균 알고리즘을 실행하여 해당 영역을 그룹화 후 정확도 측정.
      정확도: 25개의 영향 인자를 사용하여 자전거 집결지 유형 구분.
      Training: Meanshift clustering의 결과인 492개의 집결지의 랜덤하게 고른 75%의 집결지
      Test: Meanshift clustering의 결과인 492개의 집결지의 남은 25%
    - 4.	결과 해석<br>
      z_ Abn_DAY 와 z_cv 에 따라 4(k=4)개의 클러스터를 해석하여 사용 효율이 높은 그룹을 밝힘.
      클러스터의 공간 분포:  고효율 모드는 주로 중심 지역에 분포, 효율 모드 지역은 일반 비효율 지역과 고효율 지역 사이에 흩어짐
      지하철 네트워크 밀도가 지역에서는 주요 분포 모드가 비효율적
    - 5.	RFC<br>
      자전거 서비스 클러스터 유형에 영향을 미치는 영향 요인을 더 잘 이해하기 위해서 RFC로 25개으 요인의 중요도 분석.
      인구, 건물 수, 간선도로의 길이, 음식점 수가 중요한 4가지 변수.
      위의 k-means 결과로 찾은 4가지 그룹의 각 25개 요인의 영향도 비교.
